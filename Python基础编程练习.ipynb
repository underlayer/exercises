> ## No.04 
> #### 生成一个8位随机密码，要求必须包含大写、小写和数字

```python
# write your code here ...
import numpy as np
def fun2():

    Num=np.array(range(48,58))#数字
    Hchr=np.array(range(65,91))#大写字母
    Lchr=np.array(range(97,122))#小写字母
    key=np.random.randint(33, 126, 8)#随机生成8个ascii码
    containNum=False
    containHigh=False
    containLow=False
    s=''
    while ():
        for i in key:
            if i in Num:
                containNum=True
            if i in Hchr:
                containHigh=True
            if i in Lchr:
                containLow=True
        if(containNum and containHigh and containLow):
            break
    for i in range(8):
        s+=chr(key[i])
    return s
print('密码：',fun2())
print('密码：',fun2())
```

> ## No.05 
> #### 从文件（speech.txt）中读取文本，并统计其中每个字/词出现的次数

```python
# write your code here ...
import os
import jieba
with open('speech.txt', mode='rt',encoding='UTF8') as f:
    contents=f.read().strip('\n')

letters=''.join(contents).strip('\n')
print(type(contents))
words=jieba.lcut(contents)
s1=set(letters)
s2=set(words)
dict1={}
dict2={}
for item in s1:
    dict1.update({item:contents.count(item)})
for item in s2:
    dict2.update({item:contents.count(item)})
print(dict1)
print(dict2)
```

> ## No.06 
> #### 从文件（scores.csv）中读取成绩(1000个学生，8门课)，并统计每个人/每门课的最高分、最低分、平均分、总分、方差

```python
# write your code here ...
import pandas as pd
col_names = ['A', 'B', 'C', 'D', 'E',  'F',  'G', 'H', ]#定义课程名A-H
score_data = pd.read_csv('scores.csv',names=col_names)
for col_names in score_data:
    print(col_names,'课程最高分:',score_data[col_names].max())
    print(col_names,'课程最低分:',score_data[col_names].min())
    print(col_names,'课程平均分:',score_data[col_names].mean())
    print(col_names,'课程方差:' ,score_data[col_names].var())
    print('-'*8)
score_data['max']=score_data.max(axis=1)
score_data['min']=score_data.min(axis=1)
score_data['mean']=score_data.mean(axis=1)
print(score_data['max'])
print(score_data['min'])
print(score_data['mean'])
```
